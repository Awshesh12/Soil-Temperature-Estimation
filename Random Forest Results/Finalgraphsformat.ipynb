```python
from statistics import mean
import numpy as np

def best_fit_slope(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    return m
```

```python
import pandas as pd
data=pd.read_csv('predictionsRFhourlyunop.csv')
data1=data['testingvalues']
data2=data['predictedvalues']
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
SUP = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
SUB = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
a="R2".translate(SUP)
c="0C".translate(SUP)
title_font = {'fontname':'Arial', 'size':'16', 'color':'black'}
```

```python
fig, ax = plt.subplots()
xs = np.array(data1, dtype=np.float64)
ys = np.array(data2, dtype=np.float64)
m = best_fit_slope(xs,ys)
b = mean(ys) - m*mean(xs)
import matplotlib.pyplot as plt
ax.scatter(data1, data2)
ax.plot([10, 30], [10, 30], 'k--', lw=4)
y=m*data1+b
ax.plot(data1,y,color='red')
ax.set_xlabel('Measured Ts'+' ('+str(c)+')')
ax.set_ylabel('Estimated Ts'+' ('+str(c)+')')
ax.text(10.2,28.25,'RF', **title_font)
ax.text(10.2,26.25,'y='+str(round(m,4))+'x'+'+'+str(round(b,4)))
ax.text(10.2,24.25,str(a)+'='+str(round(r2_score(data1,data2),3)))
ax.text(10.2,22.25,'MAE='+str(round(mean_absolute_error(data1,data2),3)))
ax.xaxis.set_ticks(np.arange(10, 34, 4))
ax.yaxis.set_ticks(np.arange(10, 34, 4))
plt.savefig('RFhourlyunop.png',dpi=500)
from sklearn.metrics import r2_score
print("R2 score is" + str(r2_score(data1,data2))) 
plt.show()
```
